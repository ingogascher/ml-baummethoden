0. Importieren Sie alle die n√∂tigen Bibliotheken.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from matplotlib import cm
```

1. Kopieren und visualisieren (am besten mit scatter) Sie die Daten:

```python
studying = np.array([84,1323,282,957,1386,810,396,474,501,660,1260,1005,1110,1290])
result = np.array([44,97,30,51,95,51,44,41,21,40,90,83,61,92])

plt.scatter(studying, result)
```

```python
def mse(y_pred,y):
    return (y_pred-y)**2

#def sse(y_pred,y):
#    return np.mean((y_pred-y)**2)

def sse(y_pred, y):
    return np.sum((y_pred-y)**2)

for w in np.linspace(0.05, 0.09, 5):
    print(w)
```

2. Implementieren Sie das Gradientenverfahren f√ºr die Regressionsaufgabe. Es ist ausreichend, lineare Regression zu betrachten. 

```python
# Initialgewicht
w = 1
# Initialbias
b = 1
# Anzahl der Epochen:
epochs = 2000
# Learning rate:
lr_w = 0.00000001
lr_b = 0.001

#F√ºr Plots speichern wir alle "zwischendrin" Fehler und Gewichte
b_history, w_history, error_history = [],[],[]

# Vergesses Sie nicht, den Fehler bei jeder Iteration f√ºr die Visualisierung zu speichern.
for x in range(epochs):

  predicted = w * studying + b
  error = sse(predicted, result)

  gradient_w = 2 * np.dot(studying, predicted - result)
  gradient_b = 2 * np.sum(predicted - result)

  w = w - lr_w * gradient_w
  b = b - lr_b * gradient_b

  w_history.append(w)
  b_history.append(b)
  error_history.append(error)


print("Ergebnis: w: {} b:{}" . format(w,b))
```

3. Visualisieren Sie den Fehler: 

```python
w_np = np.array(w_history)
w_np = w_np.reshape(-1,1)
w_norm = MinMaxScaler().fit_transform(w_np)

b_np = np.array(b_history)
b_np = b_np.reshape(-1,1)
b_norm = MinMaxScaler().fit_transform(b_np)

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(131)
ax.plot(range(epochs), w_norm)
ax.plot(range(epochs), b_norm)

ax = fig.add_subplot(132)
ax.plot(range(epochs), w_history)
ax.plot(range(epochs), b_history)

ax = fig.add_subplot(133)
ax.plot(range(epochs), error_history)
ax.set_yscale('log')
```
