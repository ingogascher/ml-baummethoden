```python
import pandas as pd

data = {
    'studying':[84,1323,282,957,1386,810,396,474,501,660,1260,1005,1110,1290],
    'result':[44,97,30,51,95,51,44,41,21,40,90,83,61,92]
}

# Create DataFrame
df = pd.DataFrame(data)

# Print the output.
print(df)
```

```python
#df.plot.scatter(x='studying', y='result', c='DarkBlue')

from matplotlib import pyplot as plt
plt.scatter(df['studying'], df['result'])
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

def mse(y_pred,y):
    return np.mean((y_pred-y)**2)

ws = np.linspace(0, 0.1, 5)
bs = np.linspace(15,25, 5)
combinations = np.array([[w, b] for w in ws for b in bs])
x_axis = np.linspace(0, 1500)

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
ax.scatter(df['studying'], df['result'])
ax.set_title("Lineare Regression")
ax.set_xlabel("Vorbereitungszeit auf Klausur in Minuten")
ax.set_ylabel("Klausurergebnis")

for wb in combinations:
    w, b = wb[0], wb[1]
    prediction = w * df['studying'] + b
    error = mse(prediction, df['result'])
    ax.plot(x_axis, w * x_axis + b, label="(w,b) = ({}, {}), Fehler: {}".format(round(w,3),b, round(error,3)))
ax.legend()
```
