```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "/home/ingo/Projects/ml-tutorial/data/Census-Datensatz.csv", sep=';', header=0, encoding='utf-8'
)
df.head()
```

```python
# check for missing values
df.isna().sum()
df.isnull().sum()
```

```python
# check if ? is in any of the columns
for col in df.columns:
    if "?" in df[col].values:
        print(col)
```

```python
# get all unique values for the columns
for col in df.columns:
    print(df[col].unique())
```

```python
# remove all empty spaces from string values in columns
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip()
    print(df[col].unique())

```

```python
# check all columns for ? and get columns that contain ?
for col in df.columns:
    if "?" in df[col].values:
        print(col)
```

```python
# define missing value dictionary and set nan for all  these values
nan_dict = {
    "?": np.nan,
    "": np.nan
}
for col in df.columns:
    df[col] = df[col].replace(nan_dict)

df.isnull().sum()
```

```python
# get all instances where workclass is missing
df.loc[df["workclass"].isnull()]
```

```python
df.dropna()  # drop all lines with missing values
df.dropna(axis=1)  # drop all columns with missing values

df2 = df.dropna() # save clean data frame in new variable

df["workclass"].fillna("Unknown")
df.fillna({"workclass": "Unknown", "native-country": "Don't know"})  # fill missing values with

df.select_dtypes(include="object").fillna("Unknown")  # only fill nan values for object columns

```

```python
# data frame with only females
df_female = df.loc[df["sex"] == "Female"]

df.drop(index = [1, 2])  # drop index or list of index

df.drop_duplicates()  # drop all duplicated lines

# df.set_index()  # to set a specific column as index, for example patient_id

df.sort_values(by=["age"], ascending=False)
```
